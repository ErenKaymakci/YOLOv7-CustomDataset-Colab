data.yaml dosyanız yoksa endişe etmeyin zipin içerisine yaml dosyası oluşturup boş bir şekilde atın düzenlemeyi beraber yapalım.
## YAML Dosyası Düzenleme
yaml dosyasının içeriği bu şekilde olmalıdır:
```
train: ../data/train/images
val: ../data/val/images

nc: 5 
names: [...]
```
<ul type="disc">
<li> <strong>nc</strong> : sınıf sayınız </li>

<li> <strong>names</strong> : sınıf isimleriniz örn: [insan, kedi, köpek] </li>

<li>  

```
train: ../{klasör adı}/train/images 
val: ../{klasör adı}/val/images
```
Ekstra test klasörünüz varsa val satırının altına: 

``` test: ../test/images ```
 vb. şeklinde ekleyebilirsiniz 
</li>

</ul>
